```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

import pytesseract
from pytesseract import Output
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'
```

### Example 1

```python
# load image
img = cv2.imread('faktura.png')
```

```python
Image.fromarray(img)
```

```python
# get text from image using pytesseract
text = pytesseract.image_to_string(img, lang="pol")
```

```python
print(text)
```

```python
# get bounding boxes
data = pytesseract.image_to_data(img, output_type=Output.DICT)
n_boxes = len(data['level'])
for i in range(n_boxes):
    (x, y, w, h) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
Image.fromarray(img)
```

### Example 2

```python
img2 = cv2.imread('zarzadzenie.PNG')
```

```python
Image.fromarray(img2)
```

```python
# get text from image using pytesseract
text2 = pytesseract.image_to_string(img2, lang="pol")
```

```python
print(text2)
```

```python
# get bounding boxes of text only
data2 = pytesseract.image_to_data(img2, output_type=Output.DICT)
n_boxes2 = len(data2['level'])
for i in range(n_boxes2):
    if data2['text'][i] != "":
        (x, y, w, h) = (data2['left'][i], data2['top'][i], data2['width'][i], data2['height'][i])
        cv2.rectangle(img2, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
Image.fromarray(img2)
```
